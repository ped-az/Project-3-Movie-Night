```python
import pandas as pd
```

```python

df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
df.info()
```

```python
# Remove the wrong year number which contains alphabetic characters and convert year to numeric
df = df[~df['Released_Year'].str.contains('^[a-zA-Z]')]
df['Released_Year']=df['Released_Year'].astype(int)
```

```python
# Remove ' min' string from runtime and convert runtime to numeric
df['Runtime'] = df['Runtime'].str.replace(" min","")
df['Runtime']=df['Runtime'].astype(int)
```

```python
# Remove ',' string from gross and convert it to numeric
df['Gross'] = df['Gross'].str.replace(",","")
df['Gross']=df['Gross'].astype(float)
```

```python
df['Gross'].sum()
```

```python
df.info()
```

```python
df.head()
```

```python
# Split the genre to list of genres
df['Genre'] = df['Genre'].str.split(',')
# Count number of genres for each movie
df['No_of_Genres'] = df['Genre'].str.len()
df['No_of_Genres'].describe()
```

```python
# Extract the genres
df['Genre_1'] = df['Genre'].str[0]
df['Genre_2'] = df['Genre'].str[1]
df['Genre_3'] = df['Genre'].str[2]
```

```python
df.info()
```

```python
list(df.columns)
```

```python
df.columns = ['Poster', 'Title',
 'Year',
 'Certificate',
 'Runtime',
 'Genre',
 'IMDB',
 'Overview',
 'MetaScore',
 'Director',
 'Star1',
 'Star2',
 'Star3',
 'Star4',
 'Votes',
 'Gross',
 'Genres',
 'Genre1',
 'Genre2',
 'Genre3']
```

```python
df.to_csv('imdb_top_1000_cleaned.csv', index=False)
```
